```python
# method 0x434C49434B

# everytime the click points to 0 now counts 

#Day 1 Code (redesigned for unit testing - error in code):

count = 0
position = 50

nums = ['L68', 'L30', 'R48', 'L5', 'R60', 'L55', 'L1', 'L99', 'R14', 'L82']

for num in nums:
    rotation = num[0]
    step_size = int(num[1:])
    zero = False 
    if rotation == 'L':
        if step_size > position and position != 0:
            zero = True
        position -= step_size
        
    elif rotation == 'R':
        if (99 - position) < step_size:
            zero = True
        position += step_size

    position = position % 100 

    if position == 0 or zero:
        count += 1

count
```

```python
#unit testing - should return 6 

nums = ['L68', 'L30', 'R48', 'L5', 'R60', 'L55', 'L1', 'L99', 'R14', 'L82']

count = 0
position = 50

for num in nums:
    rotation = num[0]
    step_size = int(num[1:])
    
    current_clicks = 0

    if rotation == 'R':
        current_clicks = (position + step_size) // 100
        
        position = (position + step_size) % 100

    elif rotation == 'L':
        if position == 0:
            current_clicks = step_size // 100
        else:
            if step_size >= position:
                current_clicks = 1 + (step_size - position) // 100
            else:
                current_clicks = 0
                
        position = (position - step_size) % 100

    count += current_clicks

count
```

```python
#Day two final code 

count = 0
position = 50

with open('input.txt', 'r') as file:

    for line in file:
        text = line.strip()

        if not text:
            continue
            
        moves = text.split(',')

        for move in moves:
            move = move.strip()
            if not move:
                continue

            rotation = move[0]
            step_size = int(move[1:])
            
            current_clicks = 0

            if rotation == 'R':
                current_clicks = (position + step_size) // 100
                position = (position + step_size) % 100

            elif rotation == 'L':
                if position == 0:
                    current_clicks = step_size // 100
                else:
                    if step_size >= position:
                        current_clicks = 1 + (step_size - position) // 100
                    else:
                        current_clicks = 0
                
                position = (position - step_size) % 100

            count += current_clicks

count
```
