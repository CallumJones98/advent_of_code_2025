```python
# Rotation 
# Left to the lower numbers and Right to the higher numbers
# distance value indicates how many clicks the dials should be rotated in that direction
#Â l1 with a rotation of r8 would get to l9
# it is a dial left from 0 = 99
# right from 99 = 0 

# dial starts by pointing at 50 

# the password is the number of times the dial points at 0

#Example/unit test: 

nums = ['L68', 'L30', 'R48', 'L5', 'R60', 'L55', 'L1', 'L99', 'R14', 'L82']

count = 0
position = 50

for num in nums:
    rotation = num[0]
    step_size = int(num[1:])

    if rotation == 'L':
        position -= step_size
    elif rotation == 'R':
        position += step_size

    position = position % 100 
    
    if position == 0:
        count += 1


count #should return 3 
```
