```python
# batteries are arranged into banks - a line of digits 
#Â Within each battery you need to turn on exactly two batteries 
#Â e.g. if you need 24 jolts and the battery looks like 12345 you would turn on battery 2 and 4 = 24
# You cannot rearrange batteries
# You need to find the largest possible joltage each bank can produce 

total = 0

with open('input.txt', 'r') as file:
    for row in file:
        row = row.strip()
        row_split = [int(val) for val in str(row)]
        if max(row_split) != row_split[-1]:
            first_volt = max(row_split)
            second_volt = max(row_split[row_split.index(first_volt)+1::])

        elif max(row_split) == row_split[-1]:
            first_volt = max(row_split[:-1])
            second_volt = max(row_split[row_split.index(first_volt)+1::])
            
        total += int(str(first_volt) + str(second_volt))

total

                
```

```python
# Part two code

# make the largest joltage by turning on extactly twelve batteries within each bank

total = 0

with open('input.txt', 'r') as file:
    for row in file:
        row = row.strip()
        target_len = 12
        nums = str(row)
        to_remove = len(nums) - target_len
        num_li = []

        for val in nums:
            while to_remove > 0 and num_li and num_li[-1] < val:
                num_li.pop()
                to_remove -= 1

            num_li.append(val)

        final_num = num_li[:target_len]

        total += int(''.join(final_num))

total
```
