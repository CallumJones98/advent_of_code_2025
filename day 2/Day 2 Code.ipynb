```python
# ID ranges appear on one single line 
# ranges are separated by commas (,)
# each range gives its first ID and last ID separated by a dash (-)

# Invalid IDs include sequences of digits where the digit is repeated twice e.g. 55 or 6464 and 123123
# None of the numbers have leading 0's e.g. 0101 

# Find all invalid IDs that appear in the given ranges:

#Â 11-22 
# 99
# 1010
# 1188611885

with open('input.txt', 'r') as file:
    content = file.read()
    all_data = content.split(',')

    total = 0 

    for row in all_data:
        num_one, num_two = row.split('-')
        for val in range(int(num_one), int(num_two) + 1):
            if len(str(val)) % 2 == 0:
                half = len(str(val)) // 2
                first_half = str(val)[0:half]
                second_half = str(val)[half::]
                if first_half == second_half:
                    total += int(val)
                else:
                    continue

    print(total)
                
            
    
```

```python
# Part 2 code 
import re 

with open('input.txt', 'r') as file:
    content = file.read()
    all_data = content.split(',')

    total = 0 

    for row in all_data:
        num_one, num_two = row.split('-')
        for val in range(int(num_one), int(num_two) + 1):
            match = re.fullmatch(r'(.+?)\1+', str(val))
            if match:
                total += int(val)
            else:
                continue

    print(total)

```
